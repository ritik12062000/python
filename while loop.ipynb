```python
n = int(input("Enter number: "))
sum = 0
while n > 0:
    sum += n % 10
    n //= 10
print("Sum of digits:", sum)
```

```python
n = int(input("Enter number: "))
count = 0
while n > 0:
    count += 1
    n //= 10
print("Total digits:", count)
```

```python
n = int(input("Enter number: "))
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print("Reversed number:", rev)
```

```python
n = int(input("Enter number: "))
temp = n
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
n = int(input("Enter number: "))
product = 1
while n > 0:
    product *= n % 10
    n //= 10
print("Product of digits:", product)
```

```python
n = int(input("Enter number: "))
largest = 0
while n > 0:
    digit = n % 10
    if digit > largest:
        largest = digit
    n //= 10
print("Largest digit:", largest)
```

```python
n = int(input("Enter number: "))
smallest = 9
while n > 0:
    digit = n % 10
    if digit < smallest:
        smallest = digit
    n //= 10
print("Smallest digit:", smallest)

```

```python
n = int(input("Enter number: "))
temp = n
sum = 0
while n > 0:
    digit = n % 10
    sum += digit ** 3
    n //= 10

if sum == temp:
    print("Armstrong number")
else:
    print("Not Armstrong")
```

```python
n = int(input("Enter number: "))
result = 0
place = 1

while n > 0:
    digit = n % 10
    if digit != 0:
        result = digit * place + result
        place *= 10
    n //= 10

print("Number without zeros:", result)
```

```python
n = int(input("Enter number: "))
even = 0
odd = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        even += 1
    else:
        odd += 1
    n //= 10

print("Even digits:", even)
print("Odd digits:", odd)
```

```python
n = int(input("Enter number: "))
found = False

while n > 0:
    if n % 10 == 5:
        found = True
        break
    n //= 10

if found:
    print("Digit 5 is present")
else:
    print("Digit 5 not present")
```

```python
n = int(input("Enter number: "))
even_sum = 0
odd_sum = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        even_sum += digit
    else:
        odd_sum += digit
    n //= 10

print("Even digit sum:", even_sum)
print("Odd digit sum:", odd_sum)
```

```python
n = int(input("Enter number: "))
result = ""

while n > 0:
    digit = n % 10
    result = str(digit * digit) + result
    n //= 10

print("New number:", result)
```

```python
n = int(input("Enter number: "))
i = 1
sum = 0

while i < n:
    if n % i == 0:
        sum += i
    i += 1

if sum == n:
    print("Perfect number")
else:
    print("Not Perfect")
```

```python
n = int(input("Enter terms: "))
a, b = 0, 1
i = 1

while i <= n:
    print(a, end=" ")
    a, b = b, a + b
    i += 1
```
